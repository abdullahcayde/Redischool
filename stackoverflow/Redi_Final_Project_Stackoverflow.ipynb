## Create New Dataset from original dataset

```python
# use conatins , case not sensitive and ignore NAN rows
dff = df_first.loc[df_first.DevType.str.contains('data scientist|data or business analyst', case=False ,na=False)]
dff.head(1)
```

```python
dff.describe().round(2)
```

## Find Salary from Compensation 

```python
# Feaure Engineering

# New Column Compensation Frequency Number
#dff['CompFreq_number'] = dff.CompFreq.apply(lambda x : (52/12) if x=='Weekly' else 12 )

dff.loc[dff.CompFreq == 'Weekly', 'CompFreq_Number'] = (52/12)
dff.loc[dff.CompFreq == 'Monthly', 'CompFreq_Number'] = 1
dff.loc[dff.CompFreq == 'Yearly', 'CompFreq_Number'] = 12
dff.loc[dff.CompFreq == np.NaN , 'CompFreq_Number'] = np.NaN

pd.options.display.float_format = '{:,.2f}'.format
dff[['CompTotal', 'CompFreq', 'CompFreq_Number']]
```

```python
# Find Salary (Monthly income)
dff['Salary'] = dff.CompTotal / dff.CompFreq_Number
dff[['CompTotal', 'CompFreq', 'CompFreq_Number', 'Salary',  'WorkExp','ConvertedCompYearly']]
```

```python
# dff to excel

writer = pd.ExcelWriter('Stackoverflow.xlsx')

# df.info(), df.describe(), df.quantile()
dff.dtypes.reset_index(name='Dtype').rename(columns={'index' : 'Column'}).to_excel(writer, 'info')
dff.describe().T.to_excel(writer, 'describe')
dff.quantile([0.01,0.25,0.50,0.75,0.99]).T.to_excel(writer, 'quantile')

writer.save()
```
