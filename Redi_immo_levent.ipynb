```python
def outlier_equal_zero(df, col):
    q1 = df[col].describe()[4]
    q3 = df[col].describe()[6]
    intr_qr = q3 - q1 
    upper_bound = q3 + (1.5 * intr_qr)
    lower_bound = q1 - (1.5 * intr_qr)

    df.loc[df[col] < lower_bound, col] = lower_bound
    df.loc[df[col] > upper_bound, col] = upper_bound

    return df[col].max()
```

```python
outlier_equal_zero(df, 'Rooms')
outlier_equal_zero(df, 'Space')
outlier_equal_zero(df, 'Age')
df.describe(percentiles=[.01,.05,.25,.5,.75,.95,.99]).T
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

Ho = "Condition has no effect on Rent"   # Stating the Null Hypothesis
Ha = "Condition has an effect on Rent"   # Stating the Alternate Hypothesis

model = ols('Rent ~ C(Condition)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
p_value = anova_table['PR(>F)'][0]

if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value.round(3)}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)}) > 0.05')

anova_table
```
