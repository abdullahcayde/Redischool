# Anova

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

Ho = "Region has no effect on Rent Price "   # Stating the Null Hypothesis
Ha = "Region has an effect on Rent Price"   # Stating the Alternate Hypothesis

model = ols('Rent ~ C(Region)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
p_value = anova_table['PR(>F)'][0]

if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value.round(3)}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)}) > 0.05')

anova_table
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

Ho = "Condition has no effect on Rent Price "   # Stating the Null Hypothesis
Ha = "Condition has an effect on Rent Price"   # Stating the Alternate Hypothesis

model = ols('Rent ~ C(Condition)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
p_value = anova_table['PR(>F)'][0]

if p_value < 0.05:  # Setting our significance level at 5%
    print(f'{Ha} as the p_value ({p_value.round(3)}) < 0.05')
else:
    print(f'{Ho} as the p_value ({p_value.round(3)}) > 0.05')

anova_table
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols

Ho = "Condition has no effect on Rent Price "   # Stating the Null Hypothesis
Ha = "Condition has an effect on Rent Price"   # Stating the Alternate Hypothesis

model = ols('Rent ~ C(Condition)+ C(Region) + C(Condition):C(Region)', data=df).fit()
anova_table = sm.stats.anova_lm(model, typ=2)
anova_table
```

```python
from bioinfokit.analys import stat
res = stat()
res.anova_stat(df=df, res_var='Rent', anova_model='Rent ~ C(Condition)+ C(Region) + C(Condition):C(Region)')
res.anova_summary
```

```python
from statsmodels.graphics.factorplots import interaction_plot
fig = interaction_plot(x=df['Condition'], trace=df['Region'], response=df['Rent'])
plt.xticks(rotation=90)
plt.show()
```

### Test Anonva assumptions
Similar to one-way ANOVA, you can use visual approaches, Bartlett’s or Levene’s, and Shapiro-Wilk test to validate the assumptions for homogeneity of variances and normal distribution of residuals.

```python
# QQ-plot
import statsmodels.api as sm
# res.anova_std_residuals are standardized residuals obtained from two-way ANOVA (check above)
sm.qqplot(res.anova_std_residuals, line='45')
plt.xlabel("Theoretical Quantiles")
plt.ylabel("Standardized Residuals")
plt.show()

# histogram
plt.hist(res.anova_model_out.resid, bins='auto', histtype='bar', ec='k') 
plt.xlabel("Residuals")
plt.ylabel('Frequency')
plt.show()

# Shapiro-Wilk test
import scipy.stats as stats
w, pvalue = stats.shapiro(res.anova_model_out.resid)
print(f'Test Statistics (w) = {w}, P-Value = {pvalue}')
```

### Levene Test
We will use Levene’s test to check the assumption of homogeneity of variances

```python
# if you have  a stacked table, you can use bioinfokit v1.0.3 or later for the Levene's test
from bioinfokit.analys import stat 
res = stat()
res.levene(df=df, res_var='Rent', xfac_var=['Condition', 'Region'])
res.levene_summary
```
